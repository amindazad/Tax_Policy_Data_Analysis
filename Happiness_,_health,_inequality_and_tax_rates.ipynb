```
#Import pandas, numpy, seaborn and matplotlib to read , clean and vizualize data
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import statsmodels as st

#Import the datasets we are going to use
Capital_Income_Tax = pd.read_csv('https://raw.githubusercontent.com/amindazad/progressive-tax-report/master/Capital%20gains%20tax%20rates%20OECD%202019.csv', header=1)
Corporate_Income_Tax = pd.read_csv('https://raw.githubusercontent.com/amindazad/progressive-tax-report/master/Corporate%20Income%20Tax%20OECD%202019.csv')
Personal_Income_Tax = pd.read_csv('https://raw.githubusercontent.com/amindazad/progressive-tax-report/master/Personal%20Income%20Tax%202019%20OECD.csv',header=1)
Better_Life_Index= pd.read_csv('https://raw.githubusercontent.com/amindazad/progressive-tax-report/master/BLI%202019%20OECD.csv')
Happiness_index= pd.read_csv('https://raw.githubusercontent.com/amindazad/progressive-tax-report/master/World%20Happiness%20index%202019.csv')
Income_inequality = pd.read_csv('https://raw.githubusercontent.com/amindazad/progressive-tax-report/master/Income%20Inequality%20OECD.csv',header=1)
GDP = pd.read_csv('https://raw.githubusercontent.com/amindazad/progressive-tax-report/master/GDP%20PPP%20per%20Capita%20USD%20OECD%20.csv')
```

```
# Let's clean each dataset and then merge them using the country name
print(Capital_Income_Tax.shape)
Capital_Income_Tax.head()
```

```
# Starting with Capital income tax data set and sliming it down to the columns we need
Capital_Income_Tax_Cleaned = Capital_Income_Tax[['Country', 'CIT rate on distributed profit']]
Capital_Income_Tax_Cleaned
```

```
#Check out the data types 
Capital_Income_Tax_Cleaned.dtypes
```

```
#Look for NaN values 
Capital_Income_Tax_Cleaned.isnull().sum().sort_values()
```

```
#Ok, It seems we're good to go, let's move on to the next dataset
#Corporate_Income_Tax
Corporate_Income_Tax.head()
```

```
# Country column is unnamed, we need that column so bad...let's name it
Corporate_Income_Tax=Corporate_Income_Tax.rename(columns={'Unnamed: 0':'Country'})
Corporate_Income_Tax.head()
```

```
# Starting with Capital income tax data set and sliming it down to the columns we need
Corporate_Income_Tax_Cleaned = Corporate_Income_Tax[['Country', 'Combined corporate income tax rate']]
Corporate_Income_Tax_Cleaned
```

```
#Check for data types and NaN values
Corporate_Income_Tax_Cleaned.dtypes
```

```
Corporate_Income_Tax_Cleaned.isnull().sum().sort_values()
```

```
# Row 36 has a NaN and its an unrelated column, lets drop it. 
Corporate_Income_Tax_Cleaned[pd.isnull(Corporate_Income_Tax_Cleaned).any(axis=1)]
```

```
#Droping row 36 
Corporate_Income_Tax_Cleaned = Corporate_Income_Tax_Cleaned.dropna()
Corporate_Income_Tax_Cleaned.tail()
```

```
#Moving on to the next dataset
Personal_Income_Tax.head()
```

```
#sliming down to the columns we need
Personal_Income_Tax_Cleaned = Personal_Income_Tax[['Country', 'Personal income tax']]
Personal_Income_Tax_Cleaned
```

```
#Check fot data types and null values 
print(Personal_Income_Tax_Cleaned.dtypes)
Personal_Income_Tax_Cleaned.isnull().sum().sort_values()
```

```
# Moving on to the next dataset 
print(Better_Life_Index.shape)
Better_Life_Index.head()
```

```
#sliming down to the columns we need
Better_Life_Index_Cleaned = Better_Life_Index [['Country','Employment Rate',
                                                '  Years in education',
                                                'Water Quality (%)',
                                                '  Life expectancy (Years)',
                                                '  Life satisfaction Average Score']]
Better_Life_Index_Cleaned.head()
```

```
#Get the whitespace out of column names
Better_Life_Index_Cleaned=Better_Life_Index_Cleaned.rename(columns=lambda x: x.strip())
Better_Life_Index_Cleaned.head()
```

```
#Moving on to the next dataset 
print(Happiness_index.shape)
Happiness_index.head()
```

```
#sliming down to the columns we need
Happiness_index_Cleaned=Happiness_index.rename(columns={'Score':'Happiness Score'})

Happiness_index_Cleaned = Happiness_index_Cleaned [['Country or region',
                                                    'Happiness Score']]

Happiness_index_Cleaned.head()                                                
```

```
#Renaming the columns and slimming down to OECD countries
Happiness_index_Cleaned = Happiness_index_Cleaned.rename(columns={'Country or region':'Country'})
Happiness_index_Cleaned.head()
```

```
Happiness_index_Cleaned['Country'] = Happiness_index_Cleaned['Country'].replace({
    'United States of America':'United States','South Korea':'Korea',
    'Slovakia':'Slovak Republic', 'Czechia':'Czech Republic' })
Happiness_index_Cleaned.head()
```

```
Happiness_index_Cleaned.loc[Happiness_index_Cleaned['Country'] == 'Czech Republic']
```

```
# Checking for NaNs and data types 
print(Happiness_index_Cleaned.dtypes)
Happiness_index_Cleaned.isnull().sum().sort_values()
```

```
#Moving on to the next dataset 
print(Income_inequality.shape)
Income_inequality.head()
```

```
#sliming down to the columns we need
Income_inequality_Cleaned = Income_inequality [['Country','Value']]
Income_inequality_Cleaned.head()
```

```
Income_inequality_Cleaned = Income_inequality_Cleaned.rename(columns={'Value':'GINI Coefficient'})
Income_inequality_Cleaned.head()
```

```
# Checking for NaNs and data types 
print(Income_inequality_Cleaned.dtypes)
Income_inequality_Cleaned.isnull().sum().sort_values()
```

```
#Moving on to the next dataset 
print(GDP.shape)
GDP.head()
```

```
#sliming down to the columns we need
GDP_Cleaned = GDP [['Country','GDP per capita USD']]
GDP_Cleaned.head()
```

```
# Checking for NaNs and data types 
print(GDP_Cleaned.dtypes)
GDP_Cleaned.isnull().sum().sort_values()
```

```
#Lets make the final dataset by mergin all cleaned datasets 

#Starting with Taxes, to reduce the columns lets take the average of all three 
#Corporate income, Capital income and Personal income
Taxes = pd.merge(Capital_Income_Tax_Cleaned, Corporate_Income_Tax_Cleaned, on='Country', how='inner')
Taxes = pd.merge(Taxes, Personal_Income_Tax_Cleaned, on='Country', how='inner')
Taxes['Average Taxes (Corporate, Personal & Capital)'] = Taxes.mean(axis=1)
Taxes = Taxes[['Country','Average Taxes (Corporate, Personal & Capital)']]

#Lets merge taxes dataset with the remaining datasets 
Final = pd.merge(Taxes, Better_Life_Index_Cleaned, on='Country', how='inner')
Final = pd.merge(Final, GDP_Cleaned, on='Country', how='inner')
Final = pd.merge(Final, Happiness_index_Cleaned, on='Country', how='inner')
df = pd.merge(Final, Income_inequality_Cleaned, on='Country', how='inner')

print(df.shape)
df
```

```
#Check for NaN values and doubel check the data types 
print(df.dtypes)
df.isnull().sum().sort_values()
```
